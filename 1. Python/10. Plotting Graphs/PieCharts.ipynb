```python
## Setting colors, explode, labels, autopct, direction

## the graph i

labels = ["A", "B", "C", "D"]
sizes = [3,4,6,2]
colors = ["pink", "blue", "red", "purple"]
explode = [0.1, 0, 0 , 0]
plt.pie(sizes, colors = colors, labels = labels, autopct = "%.2f%%", explode = explode)
plt.title("Split Among Classes")
plt.axis("equal")
plt.show()
```
