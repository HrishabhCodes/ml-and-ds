```python
b = [1, 2, 1, 3, 5, 4, 5, 5, 4, 3, 3, 10, 7, 6, 7, 8, 8, 20, 22, 11,14]
```

```python
plt.hist(b, edgecolor = "black")

plt.show()

## it has clubbed few numbers together
```

```python
plt.hist(b, edgecolor = "black")

plt.xticks([1,2,3,4,5])

plt.show()
```

```python
nums = np.arange(23)
nums
```

```python
plt.hist(b, edgecolor = "black")  ## by default bins = 10

plt.xticks(nums)

plt.show()
```

```python
plt.hist(b, bins = 25, edgecolor = "black")

plt.xticks(nums)

plt.show()
```

```python
plt.hist(b, bins = 21, edgecolor = "black")

plt.xticks(nums)

plt.show()
```

```python
# Basic format for every bin is [a,b) except the last one [a,b]
# in the above example they are: [1,2) , [2,3) , [3,4) , ............ , [20,21) , [21, 22]
```

```python
plt.hist(b, bins = [1,3,6,10,15,20], edgecolor = "black")
## Bins:  [1,3) , [3,6) , [6,10) , [10,15) , [15, 20]
# plt.xticks(nums)
plt.axis([0, 25, 0, 10])
plt.show()
```

### bins = ( max - min ) / 2
