```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100, 120, 180, 250, 300, 370]
color = ["red", "orange", "blue", "green", "black", "purple"]
col = np.arange(len(year))
```

```python
# plt.scatter(year, salary, s = 100, c = "r")
# plt.scatter(year, salary, s = population, c = color, alpha = 0.4)
# plt.scatter(year, salary, s = population, c = col, alpha = 0.4)
plt.scatter(year, salary, s = population, c = col, alpha = 0.6,edgecolor = "black", marker = "^")

plt.title("Bubble Chart")

plt.show()
```

```python
plt.scatter(year, salary, s = population, c = col, alpha = 0.6,edgecolor = "black", marker = "^")

plt.title("Bubble Chart")

plt.text(year[2], salary[2], 100)  # ----->. but we can see 100 is overlapping with the triangle

plt.show()
```

```python
# plt.scatter(year, salary, s = population, c = col, alpha = 0.6,edgecolor = "black", marker = "^")
plt.scatter(year, salary, s = population, c = col, alpha = 0.6,edgecolor = "black", marker = "o")

plt.title("Bubble Chart")

plt.text(year[2] + 0.15, salary[2] + 0.15, population[2])
plt.text(year[4] + 0.15, salary[4] + 0.15, population[4])

plt.show()
```
