```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x2, y2, "r-.")
plt.plot(x, y)
plt.show()
```

```python
## plot x^3

x = np.array([1,2,3,4])
y = x**3
plt.plot(x,y)
plt.show()
```

```python
## plot x^3

x = np.arange(0, 5, 0.1)
y = x**3
plt.plot(x,y, "r")
plt.show()
```

```python
a = [3,4,5] ## ---> considered it as y
## takes x = [0,1,2] automatically
plt.plot(a)
plt.grid()
plt.show()
```

```python
x = np.arange(0, 5, 0.1)
y = x**3
# plt.plot(x,y, "r")
# or we can add color by this as well:
# plt.plot(x, y, color = "red", marker = "+")
plt.plot(x, y, color = "red", linewidth = 5)
plt.show()
```

##### Color, marker, linewidth, labels for axis, datasets, title, axis, gird, text

```python
x = np.arange(0, 5, 0.1)

y = x**3
y2 = x**2

plt.ylabel("y")
plt.xlabel("x")

plt.title("MatplotLib Demo")

plt.plot(x, y, color = "black", linewidth = 5, label = "x^3")
plt.plot(x, y2, color = "red", linewidth = 5, label = "x^2")

plt.legend()

# plt.axis([0,10,0,200])

plt.grid()

plt.text(2, 80, "Text", fontsize = 13)

plt.show()
```
