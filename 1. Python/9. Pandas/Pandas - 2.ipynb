### Handling NaN
1. Drop na
2. Fill na

```python
## right now we don't have 'Na' entries in our data, so let's make some 
df.iloc[1:3, 2:4] = 0
df.head() ## but we dont want 0, we want NaN. For this, we have to import NumPy.
```

```python
df.iloc[1:3, 2:4] = np.nan
df.head()
```

```python
df.describe()
```

###### Drop NaN

```python
df.dropna() ## removed rows with labels 1 and 2 but df is still unchanged
```

```python
df.dropna(inplace = True)
```

```python
df.head()
```

```python
df.reset_index(drop = True, inplace = True)
```

```python
df.head()
```

##### Fill NaN

```python
df.iloc[2:4,1:3] = np.nan
df.head()
```

```python
df[df["flower_type"] == "Iris-setosa"].describe()  # we can find the means of the columns to which the NaN elements belong
```

```python
## Going with the flow of the video, we'll just be filling it with the mean of the sw column
df.sw.fillna(df.sw.mean(), inplace = True)
df.head()
```

```python
df.pl.fillna(df.pl.mean(), inplace = True)
df.head()
```
